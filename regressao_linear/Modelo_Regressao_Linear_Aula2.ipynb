```python
import pandas as pd

df = pd.read_csv('Life Expectancy Data.csv') # Importando dados
df # visualizando

# visualizando apenas colunas
# print(df.columns)
```

```python
df2 = df[df['Country'] == 'Brazil']
df2

# contar linhas
# num_linhas = df2.count()
# print(num_linhas)
```

```python
# remover espaços inicío/fim nos nomes das colunas
df2.columns = df2.columns.str.strip()
```

```python
# visualizando colunas
df2.columns
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import r2_score

# visualizar colunas selecionadas
df2[['GDP', 'Country' ]]
```

```python
lin = linear_model.LinearRegression()
lin.fit(np.c_[df2['GDP']], np.c_[df2['Life expectancy']])
# np.c_ - função NumPy concatenar arrays 2D/colunas. 4
# Aqui, ele transforma as séries pandas df2['GDP'] e df2['Life expectancy'] em arrays,
# necessários para a função fit() scikit-learn

t0, t1 = lin.intercept_[0], lin.coef_[0][0] # Intercepto -
# ponto onde a linha de regressão cruza o eixo y quando
# todas as variáveis independentes são zero
# Armazenado em um array, intercept_[0] acessa o valor do intercepto.

# O atributo "coef_" contém os coeficientes (ou "pesos") associados às
# variáveis independentes (neste caso, o GDP).
# Como coef_ é um array, coef_[0][0] acessa o valor do coeficiente
# correspondente à variável independente GDP.
```

```python
# criando gráfico - df2.plot
df2.plot(kind='scatter', x="GDP", y='Life expectancy', alpha=1, figsize=(5,3)) # alpha=transparências dos pontos // figsize=(largura, altura)

plt.xlabel("PIB") # rotulo eixo x
plt.ylabel("Expectativa de vida") # rotulo eixo y

X=np.linspace(0, 14000, 10) # np.linspace - array X de 10 valores uniformemente
# espaçados entre 0 e 14.000 - numpy
# Pontos do PIB onde a linha de regressão será desenhada/calculada

plt.plot(X, t0 + t1*X, "r") #  "r"=red // Este comando desenha a linha de regressão no gráfico.
# t0 + t1*X: Esta expressão calcula os valores de expectativa de vida previstos pela linha de regressão para os valores de X (ou seja, diferentes valores de PIB).

# Exibindo gráfico
plt.show()
```

```python
pred = lin.predict(np.c_[df2['GDP']])
print("Coeficiente de determinação: %.2f" % r2_score(np.c_[df2['Life expectancy']], pred))
```

1. ***pred = lin.predict(np.c_[df2['GDP']])***

* Biblioteca: scikit-learn

* O método predict() é utilizado para fazer previsões baseadas no modelo de regressão linear (lin) que foi previamente treinado com os dados.

* ***np.c_[df2['GDP']]***: Esta parte do código converte a coluna 'GDP' do DataFrame df2 em uma matriz 2D, necessária para a previsão. A função np.c_[] do NumPy concatena as colunas, transformando a Série do Pandas em um array bidimensional.
O que Acontece:

* O modelo lin usa os valores de PIB (coluna 'GDP') para prever os valores de expectativa de vida correspondentes. Os resultados dessas previsões são armazenados na variável pred.

2. ***print("Coeficiente de determinação: %.2f" % r2_score(np.c_[df2['Life expectancy']], pred))***

***r2_score(...):***

* Biblioteca: scikit-learn

* A função r2_score() calcula o coeficiente de determinação R², que é uma medida estatística que indica o quão bem os dados observados são explicados pelo modelo.

* np.c_[df2['Life expectancy']]: Esta parte do código transforma a coluna 'Life expectancy' em uma matriz 2D, de forma semelhante ao que foi feito para a coluna 'GDP'.

* pred: Contém as previsões geradas pelo modelo para os valores de expectativa de vida com base no PIB.
O que Acontece:

* A função r2_score() compara os valores reais de expectativa de vida (coluna 'Life expectancy') com os valores previstos (pred) e calcula o coeficiente de determinação R²

**Resumo:**

***lin.predict(...):*** Gera previsões de expectativa de vida com base no PIB.

***r2_score(...):*** Avalia o quão bem as previsões do modelo correspondem aos valores reais de expectativa de vida.

***Saída:*** Exibe o coeficiente de determinação R², que indica a qualidade do ajuste do modelo aos dados.
