```python
import pandas as pd

df = pd.DataFrame({ 'Alunos': ['Adriana','Joabe','Marcos','Paulo'],
                    'P1': [6, 7, 3, 8],
                    'P2': [7, 5, 4, 9]
                  })
```

```python
df
```

```python
#df.to_csv('notas.csv')
#df.to_excel('notas.xlsx')
df.to_html('notas.html')
```

```python
# semelhante ao df.loc
df.query('Alunos == "Marcos"')
```

```python
df[df['Alunos'] == 'Marcos']
```

```python
# criando coluna Média com lambda (APPLY - aplica função na coluna)
df['Média'] = df.apply(lambda x: (x['P1']+x['P2'])/2, axis=1)
df
```

```python
# criando coluna Média sem lambda
def media(x):
    return (x['P1'] + x['P2'])/2
```

```python
# apply/ aplicando coluna Média criada o dataframe
df['Média1'] = df.apply(media, axis=1)
df
```

```python
# bins=2: Isso significa que o histograma será dividido em 2 intervalos
# (bins). Isso resultará em duas barras, onde cada uma mostra a contagem de valores que caem em cada intervalo
todasFatecs.hist(column='Nº de Concluintes Inscritos',bins = 2)
plt.show()

# resultado -> aproximadamente 50 Fatecs têm entre 0 e 125 concluintes inscritos
# resultado -> barra baixa - aproximadamente 5 ou menos Fatecs têm entre 125 e 250 concluintes inscritos.
```
